##**Verify Data Quality**

####Missing Values

```python
import pandas as pd

def identify_missing_values(df):
  """
  This function identifies missing values and calculates their percentage for each numerical column in a DataFrame.

  Args:
      df (pandas.DataFrame): The DataFrame to analyze.

  Returns:
      pandas.DataFrame: A DataFrame with columns 'Kolom' (column name) and 'Persentase Missing Values' (percentage of missing values).
  """

  results = []
  cols = df.select_dtypes(include=['float64', 'int64'])

  for col in cols:
    # Calculate percentage of missing values directly
    percent_missing_values = (df[col].isnull().sum() / len(df)) * 100
    results.append({'Kolom': col, 'Persentase Missing Values': percent_missing_values})

  # Create and format results DataFrame
  results_df = pd.DataFrame(results)
  results_df.set_index('Kolom', inplace=True)

  return results_df

# Example usage (assuming you have your DataFrame named 'df_file')
missing_values_df = identify_missing_values(df_file)
display(missing_values_df)
```

Pada kolom ID tidak memiliki missing values karena berfungsi sebagai pengenal unik setiap data yang menjadi ciri khas data tersebut. sedangkan pada salaryMin,salaryMax dan salaryTerm menunjukkan bahwa terdapat banyak data yang hanya berisi missing values

####Outliers Values

```python
results = []

cols = df_file.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df_file[col].quantile(0.25)
  q3 = df_file[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df_file[(df_file[col] < lower_bound) | (df_file[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df_file))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df_file = pd.DataFrame(results)
results_df_file.set_index('Kolom', inplace=True)
results_df_file = results_df_file.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df_file)
```

Pada tabel di atas menunjukan 6.72% data pada kolom id teridentifikasi sebagai outlier, pada salaryMin terdapat 2.01% dan pada salaryMax sebanyak 1.88%. hal ini menunjukan bahwa terdapat beberapa data gaji minimum maupun maximum yang memiliki perbandingan nilai jauh lebih tinggi atau jauh lebih rendah dibandingkan data lainnya. pada kolom salaryTerm menunjukkan bahwa jangka waktu gaji tidak didefinisikan pada dataset.

##**Modelling Associate**

###Pre-Processing

```python
x_final = df5.drop("salaryPeriod", axis=1)
```

Penghapusan Kolom yang kurang relevan dengan dataset.

####Normalisasi Data

```python
scaler = MinMaxScaler()

x_final_norm = scaler.fit_transform(x_final)
```

Pada kolom ini dilakukan penskalaan Min-Max pada x_final. onjek scaler akan digunakan untuk mengubah skala nilai dalam data x_final agar berada pada rentang 0 dan 1 yang kemudian akan menggunakan fit_transform() dari objek scaler. data x_final akan diubah skala sesuai dengan nilai minimum dan maksimum dari setiap fitur. hal ini akan menghasilkan x_final_norm yaitu data x_final yang telah discaling

####Validasi Silang

```python
inertia_values = []

k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_final_norm)
    inertia_values.append(kmeans.inertia_)

plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.xticks(k_range)
plt.show()
```

Pada metode elbow ini dapat dilihat bahwa nnilai optimal dari klaster adalah 4. hal ini dilakukan dengan melakukan definisilist kosong pada 'inertia_values' yang digunakan sebagai penyimpanan nilai inertia dari iterasi untuk setiap klaster. untuk rentang nilai klaster yang akan dieksplorasi adalah 2-9. setiap hasil inertia dari model akan disimpan pada 'inertia_values'. setelah selesai maka dapat dilihat hasilnya dengan plot garis dan dilihat bahwa siku atau 'elbow' pada grafik ditunjukan pada nilai klaster 4

###Build Model

####K-Means

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(x_final_norm)
```

K-Means adalah algoritma klasterisasi yang digunakan untuk mengelompokkan data menjadi klaster atau kelompok yang memiliki karakteristik atau pola yang serupa berdasarkan atribut yang dimiliki. pada kode tersebut akan dilakukan pembuatan model K-Means dengan 4 klaster menggunakan data x_final_norm. langkah pertama adalah membuat objek K-Means dengan menyebutkan jumlah klaster yang diinginkan yaitu 4 dan menyertakan oarameter yaitu 'random_state-42' yang berguna untuk mengatur keacakan dan membuat hasil yang dapat direproduksi. selanjutkannya adalah melatih model dengan skala yang diubah yaitu 'x_final_norm' agar model dapat dikelompokkan menjadi 4 berdasarkan pola yang ada.

####Hierarchical Clustering

```python
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
```

Hierarchical Clustering adalah pengelompokan data ke dalam kelompok atau klaster hierarkis berdasarkan tingkat kesamaan atau jarak antar data. pada kode tersebut, dilakukan pembuatan objek AgglomerativeClustering dengan menyebutkan jumlah klaster yang diinginkan (dalam hal ini 4 klaster) menggunakan parameter n_clusters=4. selain itu untuk menentukan jenis jarak atau affinity yang digunakan untuk mengukur kedekatan antara data, dalam hal ini menggunakan 'euclidean' untuk mengukur jarak Euclidean. Parameter linkage='ward' digunakan untuk menentukan metode penggabungan klaster yang digunakan, di mana 'ward' adalah salah satu metode yang mengoptimalkan pengurangan varian dalam klaster.

###Begins and Compare Clusters

```python
kmeans_clust = kmeans.predict(x_final_norm)
hierarchy_clust = hc.fit_predict(x_final_norm)
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# List nilai n_clusters yang akan diuji
n_clusters_list = [2, 3, 4, 5]

# Inisialisasi variabel untuk menyimpan hasil terbaik
best_n_clusters = None
best_silhouette_score = -1  # Nilai awal yang sangat rendah

# Loop untuk menguji setiap nilai n_clusters
for n_clusters in n_clusters_list:
    # Inisialisasi model Hierarchical Clustering
    hierarchy = AgglomerativeClustering(n_clusters=n_clusters)

    # Fit dan prediksi menggunakan model
    y_pred = hierarchy.fit_predict(x_final)

    # Hitung silhouette score
    silhouette_avg = silhouette_score(x_final, y_pred)

    # Perbarui nilai terbaik jika silhouette score lebih tinggi
    if silhouette_avg > best_silhouette_score:
        best_silhouette_score = silhouette_avg
        best_n_clusters = n_clusters

# Cetak nilai n_clusters terbaik dan silhouette score terbaik
print("Parameter terbaik (n_clusters):", best_n_clusters)
print("Silhouette Score terbaik:", best_silhouette_score)
```
